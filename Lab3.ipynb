```python
#use loop
def print_asterisks(rows):
     for i in range(1, rows+ 1):
         print('* '*i)

def print_asterisks_1(rows):
    for i in range(rows, 0, -1):
        print('* ' * i)

def print_asterisks_2(rows):
    for i in range(1, rows + 1):
         print('  ' * (i - 1) + ' *  ' * (rows - i + 1))

def print_asterisks_3(rows):
     for i in range(1, rows+ 1):
         print('  '*((rows-1) - i + 1)+' *  '*i)

def print_asterisks_4(rows):
     for i in range(1, rows+ 1):
         print('  '*((rows-1) - i + 1)+'* '*i)

def sum_print_asterisks(rows):
    print_asterisks(rows)
    print_asterisks_1(rows)
    print_asterisks_2(rows)
    print_asterisks_3(rows)
    print_asterisks_4(rows)
    return print_asterisks(rows) , print_asterisks_1(rows) , print_asterisks_2(rows) , print_asterisks_3(rows) , print_asterisks_4(rows)

print("USE LOOP\n")
rows = int(input("Enter the number of rows: "))
if rows < 0:
    raise ValueError("Please enter a non-negative integer.")
print("Small To Large :\n")
print_asterisks(rows)

print("\nLarge To Small :\n")
print_asterisks_1(rows)

print("\nDown pyramid :\n")
print_asterisks_2(rows)

print("\nUp pyramid :\n")
print_asterisks_3(rows)

print("\nRight half pyramid :\n")
print_asterisks_4(rows)
```
